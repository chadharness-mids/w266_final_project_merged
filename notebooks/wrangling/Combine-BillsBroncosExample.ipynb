```python
def get_last_play_text(mydate):
    try:
        txt = plays.loc[plays['UTC_Datetime'] <= mydate].values[-1][3]
    except:
        txt = 'pregame'
    return txt

def get_last_play_winpct(mydate):
    try:
        pct = plays.loc[plays['UTC_Datetime'] <= mydate].values[-1][2]
    except:
        pct = 0.5
    return pct
    #print(last_play)
```

```python
comments['homeWinPercentage'] = comments['datetime_link'].apply(get_last_play_winpct)
```

```python
comments['playtext'] = comments['datetime_link'].apply(get_last_play_text)
```

```python
%matplotlib inline
starttime = pd.to_datetime('09-24-2017 16:30')
endtime = pd.to_datetime('09-24-2017 21:00')
comments.plot(x='datetime_link',
              y='homeWinPercentage',
              xlim=(starttime,endtime),
             figsize=(15,5))
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#from nltk.sentiment.util import *

sid = SentimentIntensityAnalyzer()

def calculate_VADER(comment):
    comment = str(comment)
    ss = sid.polarity_scores(comment)
    neg = ss['neg']
    neu = ss['neu']
    pos = ss['pos']
    compound = ss['compound']
    return ss, neg, neu, pos, compound
```

```python
comments['ss'], comments['neg'], comments['neu'], comments['pos'], comments['compound'] = \
    zip(*comments['comment_body'].map(calculate_VADER))
```
